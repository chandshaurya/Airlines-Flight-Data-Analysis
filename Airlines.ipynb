# What are the airlines in the dataset, accompanied by their frequencies?

```python
data['airline'].value_counts()
```

```python
data['airline'].value_counts(ascending=True).plot.barh()
plt.title("Airline with Frequencies")
plt.xlabel("No of Flights")
plt.show()
```

# Show Bar Graphs representing the Departure Time & Arrival Time.

```python
data['departure_time'].value_counts()
```

```python
data['arrival_time'].value_counts()
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)

plt.bar(data['departure_time'].value_counts().index, data['departure_time'].value_counts().values )
plt.xlabel("Departure Time")
plt.ylabel("D. Freq")
plt.title("Departure Time")

plt.subplot(1,2,2)

plt.bar(data['arrival_time'].value_counts().index, data['arrival_time'].value_counts().values )
plt.xlabel("Arrival Time")
plt.ylabel("A. Freq")
plt.title("Arrival Time")

plt.show()
```

People prefer early morning and evening for departure.  For arrival prefer in Nights

# Does price varies with airlines ?

```python
d = data.groupby('airline',as_index=False)['price'].mean().sort_values('price',ascending=False)

d
```

Vistara Airline has highest average price . Price may be changed on Class wise . It represent Whole.

```python
sns.catplot(x='airline',y='price',data=data,kind='bar',palette='rocket', hue='class')
plt.title("Airlines with along its avg Prices")
plt.show()
```

# Does ticket price change based on the departure time and arrival time?

```python
data.groupby('departure_time')['price'].mean()
```

```python
data.groupby('arrival_time')['price'].mean()
```

```python
sns.catplot(x='departure_time', y='price', kind = 'bar',data=data , palette='rocket')

plt.title(" Departure Time vs Price ")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.catplot(x='arrival_time', y='price', kind = 'bar',data=data , palette='rocket')

plt.title(" Arrival Time vs Price ")
plt.xticks(rotation=45)
plt.show()
```

People mostly prefer Evening and Nights for Travelled.

```python
sns.relplot(x='arrival_time',y='price', data=data, col= 'departure_time',kind='line')

plt.show()
```
